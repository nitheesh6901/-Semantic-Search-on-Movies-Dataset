```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("wiki_movie_plots_deduped.csv")
```

```python
df
```

```python
df = df.dropna(subset=["Cast"])                          # Here subset remove all rows which containing NaN values
```

```python
df
```

```python
df.isnull().sum()                             # isnull().sum() is used to find the sum of null values
```

```python
df = df[["Title","Genre","Plot"]]                     # here we take df as title , genre and plot
```

```python
df
```

```python
df['Plot'] = df['Plot'].str.replace(r'\s+','',regex=True).str.strip()      # remove all the spaces, tabs and newlines
```

```python
df['Genre'] = df['Genre'].replace('unknown', '')                           # it replaces the "unknown" value with ""
```

```python
df['Text'] = df['Title'] + " " + df['Genre'] + " "+df["Plot"]              # add additional column as Text
```

```python
df.isnull().sum()
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")                                     #sentence trnasformer is used for embeddings
```

```python
movie_embeddings = model.encode(df['Text'].tolist(),show_progress_bar = True)       # encode is used to covert all text into vector 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend_movie(user_description,top_k=1):
    user_embeddings = model.encode([user_description])                            #covert user embeddings into numerical vector
    similarities = cosine_similarity(user_embeddings,movie_embeddings)[0]         # it finds the similarity between the each movie with user's query
    top_indices = np.argsort(similarities)[::-1][:top_k]                          # np.arsort() is used to sort similarities,[::-1] reverse order
    recommended_movies =  df.iloc[top_indices][["Title","Genre","Text"]].copy()   #looks for the movie details for top matches
    recommended_movies['Score'] = similarities[top_indices]                       # add similarity scores
    return recommended_movies
```

```python
user_input = "A fantasy movie about Hercules"                                    # user's input query
recommendations = recommend_movie(user_input,top_k=5)
print(recommendations)
```
